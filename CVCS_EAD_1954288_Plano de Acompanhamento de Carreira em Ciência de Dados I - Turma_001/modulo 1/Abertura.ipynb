Olá, estudante!

Boas-vindas à disciplina Carreira S/A.

Esta disciplina é composta por uma trilha de aprendizagem que proporciona reflexões sobre a importância de ter um planejamento de carreira que ajuda você a pensar de forma realista e a prever a sua trajetória profissional e suas conquistas daqui a alguns anos.

Além isso, apresenta o conceito e diferença entre empreendedorismo e intraempreendedorismo, abordando as características mais atuais que são tendência no mercado de trabalho.

Vamos iniciar nossa jornada?

Temas Abordados
Estudante, nesta área você encontrará os Módulos de Aprendizagem e seus respectivos Materiais Didáticos.

Seu curso está dividido em quatro módulos de aprendizagem, sendo eles:

Módulo I – introdução à Carreiras;
Módulo II – Planejamento de Carreira;
Módulo III – Empreendedorismo;
Módulo IV – Intraempreendedorismo.
Acesse cada módulo para conferir os conteúdos.

Mapa da Disciplina
Interaja com o recurso e saiba mais sobre os itens e informações sobre o Programa de Acompanhamento de Carreira.








**MAIS DICAS**

Organize seus estudos de maneira que passem a fazer parte da sua rotina. Por exemplo, você poderá determinar um dia e horário fixos como o seu “momento do estudo”;

Procure se alimentar e se hidratar quando for estudar, lembre-se de que uma alimentação saudável pode proporcionar melhor aproveitamento do estudo;

No material de cada Unidade, há leituras indicadas. Entre elas: artigos científicos, livros, vídeos e sites para aprofundar os conhecimentos adquiridos ao longo da Unidade. Sugestões de conteúdo extra no item Material Complementar, que ampliarão sua interpretação e auxiliarão no pleno entendimento dos temas abordados;


Após o contato com o conteúdo proposto, participe dos debates mediados em fóruns de discussão, pois irão auxiliar a verificar o quanto você absorveu de conhecimento, além de propiciar o contato com seus colegas e tutores, o que se apresenta como rico espaço de troca de ideias e aprendizagem.
