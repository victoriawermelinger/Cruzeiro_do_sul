**Apresentao do Mdulo**

Definir o que nos torna felizes no  uma tarefa das mais fceis. Primeiro porque so muitas as variveis que compem a produo de nossa felicidade. Segundo porque boa parte dos estudos indica que os elementos subjetivos so determinantes para nossa apreenso do que seja felicidade. E terceiro porque a nossa estrutura psquica, em contato com a dinmica social, estabelece pontos contraditrios que dificultam e tornam problemticas as noes de felicidade. 

Sendo assim, nosso terceiro mdulo apresenta essas questes como indagaes, tentando possibilitar uma reflexo sobre como a felicidade se apresenta como conceito para a Psicologia, Neurocincia e reas afins; e como a felicidade e suas problemticas se estruturam no interior da vida social, psquica e neurolgica. Para sustentar essas premissas, utilizamos trs artigos que apresentam abordagens diferentes, porm complementares, sobre o tema. 

O primeiro artigo apresenta uma reviso sobre o conceito de felicidade do ponto de vista da psiquiatria clnica. Esse artigo  resultado de uma pesquisa sobre literatura especializada no tema e chega  concluso de que a felicidade  um fenmeno altamente subjetivo, sendo mais influenciada por elementos psicolgicos e socioculturais. Vale a pena conferir e relacionar com suas caractersticas de personalidade e relacionamento.

O segundo artigo nos apresenta uma perspectiva analtica sobre os problemas da busca da felicidade em nossa sociedade contempornea. Os autores desse trabalho se fazem valer de um vasto campo terico, Psicanlise, Sociologia, Psiquiatria e Antropologia para estudar as dificuldades dos processos de felicidade nos dias atuais. 

O terceiro e ltimo artigo nos apresenta uma importante base de estudos da neurocincia e suas conexes com a prtica docente. Com base nesse trabalho,  possvel estabelecer paralelos entre a Pedagogia e a Neurocincia tendo a felicidade como questo importante, que perpassa essas duas instncias, ou seja, como um elemento vivo de nossas possibilidades de aprendizado.
