

Chegamos ao módulo 3! Vamos nos aprofundar mais um pouco sobre a felicidade? Neste módulo, você irá conhecer diferentes perspectivas sobre a felicidade, além de perceber as contradições dos processos de produção da felicidade e, por fim, refletir sobre problemas e potenciais da felicidade na sociedade contemporânea.



**Apresentação do Módulo**

Definir o que nos torna felizes não é uma tarefa das mais fáceis. Primeiro porque são muitas as variáveis que compõem a produção de nossa felicidade. Segundo porque boa parte dos estudos indica que os elementos subjetivos são determinantes para nossa apreensão do que seja felicidade. E terceiro porque a nossa estrutura psíquica, em contato com a dinâmica social, estabelece pontos contraditórios que dificultam e tornam problemáticas as noções de felicidade. 

Sendo assim, nosso terceiro módulo apresenta essas questões como indagações, tentando possibilitar uma reflexão sobre como a felicidade se apresenta como conceito para a Psicologia, Neurociência e áreas afins; e como a felicidade e suas problemáticas se estruturam no interior da vida social, psíquica e neurológica. Para sustentar essas premissas, utilizamos três artigos que apresentam abordagens diferentes, porém complementares, sobre o tema. 

O primeiro artigo apresenta uma revisão sobre o conceito de felicidade do ponto de vista da psiquiatria clínica. Esse artigo é resultado de uma pesquisa sobre literatura especializada no tema e chega à conclusão de que a felicidade é um fenômeno altamente subjetivo, sendo mais influenciada por elementos psicológicos e socioculturais. Vale a pena conferir e relacionar com suas características de personalidade e relacionamento.

O segundo artigo nos apresenta uma perspectiva analítica sobre os problemas da busca da felicidade em nossa sociedade contemporânea. Os autores desse trabalho se fazem valer de um vasto campo teórico, Psicanálise, Sociologia, Psiquiatria e Antropologia para estudar as dificuldades dos processos de felicidade nos dias atuais. 

O terceiro e último artigo nos apresenta uma importante base de estudos da neurociência e suas conexões com a prática docente. Com base nesse trabalho, é possível estabelecer paralelos entre a Pedagogia e a Neurociência tendo a felicidade como questão importante, que perpassa essas duas instâncias, ou seja, como um elemento vivo de nossas possibilidades de aprendizado. Oferece-nos oportunidade de relacionar aos vários aspectos da vida, sendo você docente ou não. Mergulhe e aproveite!
