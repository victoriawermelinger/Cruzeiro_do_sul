**Apresentao do Mdulo**

Neste segundo mdulo, estudaremos as bases cientficas da felicidade e da qualidade de vida. Para tanto, nos debruaremos em autores e autoras de diversas reas do pensamento, de modo que a compreenso sobre a felicidade e a qualidade de vida possam ser ampliadas. Desse modo, as investigaes passam por vrios pontos de vistas, s vezes complementares, outras contraditrios, abrindo assim a perspectiva para reflexes diversas, para tomadas de decises e atitudes consequentes.

No primeiro movimento de estudo, recorremos a uma breve definio multifacetada de felicidade a partir de algumas referncias filosficas. De Aristteles a Bachelard, passando por Kant e Sarte,  possvel perceber, sinteticamente, como a ideia de felicidade  muito diferente para cada filsofo, e como essas diferenas podem ser produtivas e indutoras de associaes e articulaes para vocs, estudantes. 

Num segundo momento, apresentamos por meio do artigo tica da Felicidade em Aristteles, alguns pontos chaves do pensamento aristotlico. A importncia que o filsofo grego dava a felicidade era tamanha, que ele a considerava o bem supremo, e, portanto, uma busca tica primordial. Nesse sentido, a qualidade de vida se vincula a esse caminho de conquista da felicidade plena. 

J em nossa terceira referncia, temos o artigo O conceito de felicidade em Freud, no qual a autora Jaqueline Feltrin Inada elabora como a teoria de Sigmund Freud opera com o conceito de felicidade. Apesar de trabalhar com a perspectiva de que a humanidade busca a felicidade, diferentemente da felicidade plena aristotlica, como uma necessidade para o alcance do bem supremo, a ideia de felicidade aparece como um problema para Freud. Essa busca, para o terico da psicanlise,  incessante, e combina contraditoriamente, na era moderna da sociedade ocidental, prazer e desprazer.

E no ltimo artigo de nosso mdulo, estudaremos a felicidade e o bem-estar na viso da psicologia positiva. Se em Freud a felicidade aparece como problema, aqui a inteno  tentar associar felicidade com bem estar. Mesmo com a dificuldade de traar contornos precisos sobre a felicidade, dada sua dimenso subjetiva, o artigo procura por meio de uma pesquisa com alunos universitrios dimensionar aspectos positivos de suas vidas, aspectos que de alguma forma se vinculam com a felicidade e proporcionam bem-estar.
