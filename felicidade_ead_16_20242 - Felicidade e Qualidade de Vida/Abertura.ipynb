# **Felicidade e Qualidade de Vida**



O Curso **“Felicidade e Qualidade de Vida”** apresenta algumas abordagens sobre o conceito de felicidade, bem como sua transformação ao longo do tempo e sua conexão com diversas áreas do pensamento e da sociabilidade humana. Como o título sugere, a temática da felicidade é abordada, nesse caso, em relação a perspectivas e buscas para um projeto de qualidade de vida. Sendo assim, a análise do conceito não se limita a especulações abstratas e procura ressonância nas práticas sociais, tendo a vida e seus desdobramentos como bases fundamentais do estudo. Esta trilha oferece uma reflexão sobre o sentido da vida em seus aspectos práticos e cotidianos. 

Os estudos teóricos trabalhados aqui apontam a felicidade como um processo, ou seja, apresenta uma transmissão variada de diversas orientações que estimulam uma busca, e que não estão vinculadas exclusivamente a um objetivo específico ou um fim em si mesmas. Como se costuma dizer, **a felicidade é mais como um percurso do que necessariamente como um ponto de chegada**. No entanto, não se descartam os objetivos pontuais, as necessidades concretas e imediatas desse caminho. A medida e o equilíbrio dessa balança são pontos importantes para serem observados e analisados. 

Para melhor entendimento dos pontos, o curso apresenta autores que investigaram o conceito de felicidade e suas aplicações, tendo em vista bases científicas de validação. Essa sustentação teórica é de grande importância para que a reflexão sobre os processos da felicidade ganhe densidade e aprofundamento. Nesse sentido, as referências vão desde a filosofia grega até os estudos científicos contemporâneos, que incluem o avanço da compreensão da Neurociência, em especial sobre a formação do pensamento, o desenvolvimento das competências socioemocionais e influências na tomada de decisão e escolhas realizadas durante a vida. 

Os estudos também apontam a felicidade como recurso coletivo, e não apenas como um desejo individual. Com isso, é ressaltada a necessidade de a felicidade ser encarada como projeto social, no qual o Estado e as instituições competentes têm o dever de desenvolver, tendo como horizonte a qualidade de vida da sociedade. Legislações, campanhas, investimentos, debates públicos são alguns dos exemplos de ações que tem o propósito de afirmação da felicidade como bem coletivo. 

Por fim, a felicidade se estabelece como princípio vinculado à educação e aos projetos pedagógicos de uma sociedade em duas medidas. Na primeira, a felicidade aparece como objeto que pode ser analisado, como conceito que ultrapassa seus próprios limites e adquire importância social, tendo, portanto, o conhecimento de suas estruturas e potências como fontes de interesse público. E na segunda, a felicidade se relaciona intimamente com a produção do saber, apostando no conhecimento como agente fundamental do processo de felicidade, criação e a busca pela qualidade de vida.
