# **Felicidade e Qualidade de Vida**



O Curso**Felicidade e Qualidade de Vida**apresenta algumas abordagens sobre o conceito de felicidade, bem como sua transformao ao longo do tempo e sua conexo com diversas reas do pensamento e da sociabilidade humana. Como o ttulo sugere, atemtica da felicidade  abordada, nesse caso, em relao a perspectivas e buscas para um projeto de qualidade de vida.Sendo assim, a anlise do conceito no se limita a especulaes abstratas e procura ressonncia nas prticas sociais, tendo a vida e seus desdobramentos como bases fundamentais do estudo. Esta trilha oferece uma reflexo sobre o sentido da vida em seus aspectos prticos e cotidianos. 

Os estudos tericos trabalhados aqui apontam a felicidade como um processo, ou seja, apresenta uma transmisso variada de diversas orientaes que estimulam uma busca, e que no esto vinculadas exclusivamente a um objetivo especfico ou um fim em si mesmas. Como se costuma dizer, **a felicidade  mais como um percurso do que necessariamente como um ponto de chegada**. No entanto, no se descartam os objetivos pontuais, as necessidades concretas e imediatas desse caminho. A medida e o equilbrio dessa balana so pontos importantes para serem observados e analisados. 

Para melhor entendimento dos pontos, o curso apresenta autores que investigaram o conceito de felicidade e suas aplicaes, tendo em vista bases cientficas de validao. Essa sustentao terica  de grande importncia para que a reflexo sobre os processos da felicidade ganhe densidade e aprofundamento. Nesse sentido, as referncias vo desde a filosofia grega at os estudos cientficos contemporneos, que incluem o avano da compreenso da Neurocincia, em especial sobre a formao do pensamento, o desenvolvimento das competncias socioemocionais e influncias na tomada de deciso e escolhas realizadas durante a vida. 

Os estudos tambmapontama felicidade comorecurso coletivo, e no apenas como um desejo individual. Com isso, ressaltada anecessidade de a felicidade ser encarada como projeto social, no qual o Estado e as instituies competentes tm o dever de desenvolver, tendo como horizonte a qualidade de vida da sociedade. Legislaes, campanhas, investimentos, debates pblicos so alguns dos exemplos de aes quetem opropsito de afirmao da felicidade como bem coletivo. 

Por fim, a felicidade seestabelececomo princpio vinculado  educaoeaos projetos pedaggicos de uma sociedade em duas medidas. Na primeira, a felicidade aparece como objeto que pode ser analisado, como conceito que ultrapassa seus prprios limites e adquire importncia social, tendo, portanto, o conhecimento de suas estruturas e potncias como fontes de interesse pblico. E na segunda, a felicidade se relaciona intimamente com a produo do saber, apostando no conhecimento como agente fundamental do processo de felicidade, criao e a busca pela qualidade de vida.
