### Introduo 

Repetindo a pergunta ttulo: "a felicidade existe?". Para os mais diversos interlocutores, a resposta muitas vezes seria "sim", no entanto, se mudamos a pergunta para "Voc  feliz?"  possvel que a nova questo cause um desconforto duplo: primeiro porque aquele que ouve a pergunta no sabe a resposta, afinal, no tem clareza sobre o que poderia ser "felicidade", ento no consegue responder ao questionamento. Em segundo lugar, talvez haja na sociedade atual uma certa expectativa de felicidade, e quem ouve o questionamento sente-se incomodando de no atender s expectativas de seu grupo social, ento responde um "sim" acanhado, na expectativa de encerrar o assunto.

O ponto  que em filosofia dificilmente "encerramos" um assunto. A "busca pela sabedoria" est na origem da prpria palavra [Amor (Filos) + Sabedoria (Sofia)]. A busca dos antigos filsofos gregos foi por entender como o mundo funcionava. Ainda na Grcia Antiga, a emblemtica figura de Scrates (por volta de 470 a.C. a 399 a.C.) incluiu com veemncia o ser humano no centro da discusso filosfica. A inscrio no templo de Apolo, Conhece a ti mesmo, seria deste ento um assunto recorrente entre os pensadores que viriam depois. Aristteles (385 a.C.  323 a.C.), por exemplo, ao especular sobre temas diferentes, ao discutir sobre a tica, no deixou de incluir a felicidade como tema: "[...] entendemos a felicidade considerando-a, alm disso, o mais desejvel de todas as coisas, sem cont-la como um bem entre outros."1 O pensamento aristotlico tem uma ideia de felicidade muito prxima ao que hoje chamamos de "autorrealizao", pois para o velho pensador, alcanar a felicidade ligada era a ideia de viver de forma virtuosa, e, por virtude, ele compreendia a realizao plena do potencial das coisas; como o ser humano consigo um potencial que lhe  prprio, sua capacidade de raciocinar. Assim, desenvolver essa racionalidade abrir o caminho para sermos capazes de viver de forma mais sbia, mais virtuosa, e segundo ele, mais feliz. 

**1 ARISTTELES. tica a Nicmaco. (Os pensadores). Trad. Leonel Vallandro, Gherd Bornheim. So Paulo: Abril, 1979, p. 55.**

Entre os pensadores gregos  e entre outros muitos filsofos posteriores  a Razo  a grande guia para os seres humanos; a forma como nos relacionamos com a sociedade, como pessoas ao nosso redor, a natureza, e at com ns, mesmos tudo  com maior ou menor intensidade, a depender do filsofo  seria explicado e compreendido pelo nosso poder de pensar. Alis, essa capacidade no  s nos distingue e na guia, como seria a fonte para o nosso empoderamento, autossuficincia, liberdade, e, por que no? Uma Felicidade.

Encontramos algumas dessas preocupaes presentes em linhas de pensamento como o estoicismo e o epicurismo. Afinal, se nos deparamos com a felicidade como um enigma a se solucionar, como algo a ser buscado, ou ainda, algo a ser cultivado, isso indica que para os filsofos a Felicidade no  algo to evidente, qui, espontneo, talvez ela esteja onde ainda no est, da a busca por entender que ela , o que h a ser buscado , um ser encontrado...

Estoicos e epicuristas esto entre os grupos de pensadores que mais se debruaram sobre o tema (no apenas este,  claro), produziram textos em tom de aconselhamento, no uma frmula fechada em que atitude O necessariamente atinge a felicidade B, mas um reflexo sobre o cotidiano vivido; reflexo que se no nos satisfizer enquanto aquilo que esperamos para o caminho da felicidade, mostra-nos uma razoabilidade de atitude frente s questes cotidianas, por exemplo:

Nunca devemos esquecer que o futuro no  nem totalmente nosso, nem totalmente no nosso, para no sermos obrigados a esper-lo como se estivesse por vir com toda a certeza, nem nos desesperarmos como se no estivesse por vir jamais. (EPICURO, 2002, p. 33)2

**2 EPICURO. Carta sobre a felicidade:  (a Meneceu); Trad. e apres. lvaro Lorencini e Enzo Del Carratore. So Paulo: Editora UNESP, 2002.**

Abordagens nessa direo trazem consigo uma ideia de ajuste das expectativas sobre aquilo que controlamos (que depende de ns) e aquilo que no controlamos (que depende de fatores externos: outras pessoas, leis, economia etc.). Esse ponto  que tambm  encontrado no pensamento estoico  por vezes  mal compreendido, como se fosse um autocondicionamento para negar os problemas existentes, algo como: Se no considero mais aquilo um problema, ento **ele deixa de existir** , quando na verdade esse tipo de anlise racional do cotidiano  mais sutil. No se trata de negar a existncia das coisas que nos deixam tristes ou contrariados, no  negar a existncia da dor. A questo  fazer um reexame sobre quais so as **nossas reais prioridades**, e qual valor daremos a determinados temas, ideias, aes ou coisas. Da tambm a ideia de **durao** dada  Felicidade. Vrios pensadores que meditaram sobre o tema, descartaram a possibilidade de que coisas passageiras, alegrias momentneas (bens de consumo?)  mesmo em grande quantidade  fossem por si s um sinnimo de vida feliz.

Ento, como em filosofia no costumamos dar certos assuntos como encerrados, voltemos s perguntas iniciais: A Felicidade existe? Se a Felicidade  mais do que a soma dos nossos momentos de alegria, a tristeza far parte dela? A tristeza compe a felicidade na medida em que d contexto s nossas alegrias? A Felicidade vem com a Liberdade? Com a autossuficincia? O que essas coisas significam? E ainda, quanto mais nos conhecemos, somos mais felizes?

Em filosofia, perguntas tendem a gerar novas perguntas, talvez por isso a reflexo filosfica tem tanta dificuldade em encerrar assuntos, pois, sempre h um outro caminho a se considerar.

Curiosamente, isso pode tambm ter motivado filsofos a defender que o prprio exerccio do filosofar (perguntar, questionar, buscar, etc.) seja um caminho para a Felicidade...
