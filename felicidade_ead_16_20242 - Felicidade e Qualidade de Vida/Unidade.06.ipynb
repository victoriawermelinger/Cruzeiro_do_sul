**Apresentao do Mdulo**

Chegamos ao ltimo mdulo de nossa disciplina. At aqui estudamos diversas perspectivas sobre a felicidade e o bem-estar, apoiados por uma base terica consistente e mltipla. A filosofia, a psicologia, a psicologia positiva, a psicanlise, a antropologia, as neurocincias, a pedagogia e outras reas do pensamento construram pontes entre si e alargaram as possibilidades de reflexo, tendo como fio condutor a felicidade e seus desdobramentos. 

 Neste ltimo mdulo, apresentamos reflexes e estudos sobre prticas que tm o bem estar e a felicidade como princpio ativo, como horizonte norteador e produtor de sentidos. A investigao das experincias concretas ajuda a compreender como algumas realizaes e dinmicas sociais impactam diretamente nas percepes sobre felicidade e bem-estar. Para tanto, escolhemos trs textos de suporte.

 O primeiro deles, intitulado Bem-estar profissional de professores de escolas pblicas e privadas, promove uma reflexo a partir do contato com 108 professores do ensino bsico, sendo 54 do ensino pblico e 54 do ensino privado. Essa pesquisa, amparada pela teoria das representaes sociais, aponta que em ambos os grupos, o respeito e a valorizao so fundamentais para o bem-estar docente. Contudo, houve divergncia entre os grupos no que diz respeito aos elementos que caracterizam a satisfao profissional. Esse estudo  importante, portanto, para que entendamos os critrios essenciais para a busca pelo bem estar docente e que consigamos perceber as necessidades desse processo de acordo com suas especificidades. 

 Influncia da prtica de atividade fsica sobre a qualidade de vida de usurios do SUS  o segundo texto de nosso mdulo e, como o ttulo sugere, a pesquisa se debrua sobre a relao entre a atividade fsica e a promoo da qualidade de vida em pessoas atendidas nas Unidades Bsicas de Sade da Famlia, no municpio de Campo Grande (MS). Foram entrevistadas 90 pessoas com mais de 40 anos, e os resultados apontam que o grupo de praticantes de atividade fsica apresenta, significativamente, mais caractersticas relacionadas ao processo de qualidade de vida. Neste sentido, o estudo refora a necessidade de que programas de promoo da sade estimulem e construam condies para que a populao consiga ter acesso s atividades fsicas.  

 O nosso terceiro e ltimo texto se chama Arte e humanizao das prticas de sade em uma Unidade Bsica e situa seus estudos na sala de espera de uma Unidade Bsica de Sade. A pesquisa reflete como a arte teria a capacidade de criar outros sentidos para o ambiente de espera do atendimento bsico, que em geral  desgastante, tedioso e aflito. A arte aparece aqui, ento, como um agente transformador daquela realidade, como uma importante experincia que abre perspectivas para criao de um espao mais humanizado e, por consequncia, mais saudvel e feliz. O artigo nos convoca, portanto, a pensar o papel da arte na sociedade e seu trao fundamental de produo de bem estar e reinveno da vida.
