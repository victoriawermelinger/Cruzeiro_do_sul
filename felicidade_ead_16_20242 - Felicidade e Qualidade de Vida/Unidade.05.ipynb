Neste mdulo, voc ir conhecer caractersticas do mundo VUCA e os impactos causados por ele nas relaes atuais de trabalho. **Apresentao do Mdulo**


O que  ser feliz profissionalmente?

Felicidade  um termo subjetivo, o que sugere que ele conta com a opinio de quem responde. Mas, ainda assim, podemos sugerir algumas respostas a essa pergunta: fazer algo que faa sentido para voc, ter bons relacionamentos, sentir-se valorizado(a), ter boas condies fsicas e psicolgicas para realizar o trabalho, saber manejar bem o estresse e a ansiedade decorrentes das constantes exigncias e mudanas so bons exemplos. 

Este mdulo tem a inteno de apresentar-lhe as competncias para o profissional do sculo XXI, trazer os conceitos de lifelong learner (aprendiz por toda a vida), qualidade de vida e felicidade no trabalho, alm dos conceitos de estresse, ansiedade e resilincia, para que voc possa, ao fim de seus estudos, ter aptido para identificar fatores que possam elevar o seu nvel de satisfao com sua vida profissional.

Bons estudos!
