**Indicações de Filmes**

Filmes para contemplar a felicidade:
A seguir, listamos alguns filmes com o propósito de trazer leveza para a experiência de assisti-los, tanto ao lado da família, quanto para contemplá-los sozinho(a).

- Yesterday – 
Depois de um acidente, um jovem compositor desperta numa realidade paralela onde somente ele se lembra (ou tem o conhecimento) dos Beatles e todas suas canções. Com isso ele testa o preço da fama ao tomar crédito pelas músicas ao tocá-las em seus shows.

- Pequena Miss Sunshine – 
Quando a filha caçula de uma família desajeitada é convidada para participar de um concurso de beleza para meninas pré-adolescentes, todos deixam suas diferenças de lado e se unem para atravessar o país em três dias numa Kombi amarela enferrujada.

- Pequena Grande Vida – 
Uma nova oportunidade de vida é possível quando as pessoas descobrem um método de reduzir o próprio tamanho para uma versão minúscula, a fim de terem menos gastos vivendo em pequenas comunidades que se espalham pelo mundo. Todo esse processo faz um homem redescobrir a significado que buscava para sua vida.

- O Fabuloso Destino de Amélie Poulain – 
Ao encontrar sentido na vida em ajudar (de forma criativa) as pessoas que a rodeiam através de pequenos gestos, uma jovem inocente e cheia de manias se depara com a dificuldade de correr atrás da versão perfeita do seu próprio mundo particular.

- Monty Python e o Sentido da Vida – 
O terceiro longa-metragem do grupo de humor Monty Python apresenta uma série de esquetes sobre os estágios da vida, do milagre do nascimento até a morte. A ironia, a crítica social e o bom humor, é claro, marcam presença em cada segmento do filme.

- Forrest Gump – 
Por de trás da visão peculiar de um jovem, presenciamos histórias reais dos Estados Unidos em versões criadas pela sua mente característica. Apesar das cenas cômicas apresentadas, o filme emociona ao abrir espaço para uma reflexão sobre a vida e a humanidade.

- Divertida Mente – 
Essa animação nos apresenta de forma engenhosa as emoções dentro da mente de uma adolescente, que tem que lidar com as mudanças recorrentes da vida. A líder das emoções é a Alegria, que faz de tudo para que a garota esteja sempre feliz, mas acaba entendendo melhor o papel de cada uma das emoções, incluindo a Tristeza.

- A Fantástica Fábrica de Chocolate – 
O excêntrico dono da maior fábrica de doces do planeta decide realizar um concurso mundial para escolher um herdeiro para seu império. Cinco crianças de sorte encontram um convite dourado e com isso ganham uma visita guiada pela lendária fábrica de chocolates. Apesar de muitos preferirem o primeiro filme, o remake de Tim Burton explora a estética fantasiosa que a história do autor Roald Dahl merece.

- Atypical – 
Como bônus, indicamos essa série que nos apresenta um jovem autista que está em busca de sua própria independência. Nesta jornada, repleta de desafios, ele e sua família aprendem a lidar com as dificuldades da vida e descobrem que o significado de "ser um pessoa normal" não é tão óbvio assim.

**Filmes para refletir sobre a felicidade:**
Não que os filmes para contemplar a felicidade carecem de densidade para refletirmos sobre a vida, e nem que as produções a seguir não possam ser encaradas como puro entretenimento, mas apesar das histórias listadas aqui não nos trazerem primeiramente o sentimento de felicidade, ao assisti-las, certamente ficarão rondando a nossa mente ao nos fazerem questionar pontos específicos da nossa jornada.

- Na Natureza Selvagem – 
Depois de cumprir os deveres impostos pela vida regrada que a educação dos pais o deram, um jovem decide explorar sua liberdade viajando pelo país e vivendo experiências transformadoras que ao fim, trazem a mensagem mais importante de sua vida.

- Comer Rezar Amar – 
Certo dia, uma mulher que acaba de passar por problemas no relacionamento, decide largar tudo e viver novas experiências em lugares diferentes por um ano inteiro. Em três diferentes países, ela se reencontrará e uma grande viagem de autoconhecimento.

- Capitão Fantástico – 
Os pais de seis filhos decidem cria-los de uma forma não convencional, com o propósito da autossuficiência. Mas um triste acontecimento leva a família a deixar o isolamento e reencontrar parentes distantes, trazendo à tona velhos conflitos.

- Brilho Eterno de uma Mente sem Lembranças – 
Com o fim do relacionamento, ela se submete a um experimento para tirar seu ex-parceiro de todas as suas memórias. Apesar dele ainda nutrir a paixão pela ex-parceira, também acaba cedendo ao experimento de apagá-la da mente, mas após revisitar as memórias construídas ao lado dela, ele se arrepende e inicia uma corrida contra a própria mente.

- À Procura da Felicidade – 
Um homem de família enfrenta sérios problemas financeiros e apesar de todas as tentativas em manter a família unida, passa a cuidar sozinho de seu filho de apenas 5 anos. Entre várias dificuldades, a esperança de prover uma vida melhor para a criança traz a superação e a persistência como a maior força da mensagem encontrada nessa história.

- BoJack Horseman – 
Como bônus desta lista, indicamos uma animação que certamente será questionada no propósito de estar aqui, por nos apresentar um personagem decadente que disfarça sua baixa autoestima com uísque e relações fracassadas. Mas por trás dessa estrutura densa, há muito o que tirar daquilo que realmente importa nas nossas vidas para a plenitude da felicidade.
